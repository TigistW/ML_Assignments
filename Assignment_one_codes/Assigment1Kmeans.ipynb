```python
# Import packages
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Data visualization
data = np.array([[1, 21], [2, 65], [3, 5], [4, 25], [5, 60], [6, 15],
 [7, 10], [8, 60], [9, 40], [10, 35], [11, 48], [12, 20], [13, 40],
[14, 25], [15, 34], [16, 22], [65, 17], [40, 18], [75, 19], [70, 20], 
[60, 21], [70, 22], [55, 23], [50, 24], [10, 25], [5, 26], [10, 27],
[75, 28], [20, 29], [20, 30], [79, 31], [51, 32], [53, 33], [59, 34], 
[74, 35], [59, 36], [73, 37], [57, 38], [69, 39], [75, 40], [51, 41],
[32, 42], [40, 43], [47, 44], [53, 45], [58, 46], [59, 47], [7, 48]])

# Initialize the object
# Change the `n_clusters` to k=2 or k=3 or k=4
for n in [2, 3, 4]:
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(data)
    print("The cenroids when the cluster is", n, "is")
    print(kmeans.cluster_centers_)
    plt.scatter(data[:,0],data[:,1], c=kmeans.labels_, cmap='rainbow')
    plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='black')
    plt.show()

```

To determine the best value of k, the elbow method is often used. In the elbow method, the value of k is increased and the sum of squared distances of samples to their closest cluster center is calculated for each value of k. The value of k at which the sum of squared distances begins to decrease at a slower rate is selected as the best value of k.

In this case, we can conclude that the best value of **k is 2** based on the elbow method and the scatter plots.
